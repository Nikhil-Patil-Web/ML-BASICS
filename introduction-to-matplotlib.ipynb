## Two options for subplots

```python
#Option 1

fig, ((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))

#Plot data to each different axes

ax1.plot(x,x/2);
ax1.set(title="Striaght Line Graph");
ax2.scatter(np.random.random(10), np.random.random(10));
ax2.set(title="Scattered Graph");
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax3.set(title="Bar Graph")
ax4.hist(np.random.randn(1000));
ax4.set(title="Histogram");
```

```python
# Subplots Option 2
fig,ax=plt.subplots(nrows=2, ncols=2,figsize=(10,5));
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

## Plotting a histogram from a dataframe

```python
heart_disease=pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
heart_disease["age"].plot.hist();
```

```python
heart_disease["age"].plot.hist(bins=10);
```

```python
heart_disease.head()
```

```python
heart_disease.plot.hist(figsize=(10,30), subplots=True);
```

```python
#How do we determine the method we need to use?

over_50=heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
over_50.plot(kind='scatter',x='age',y='chol',c='target');
```

```python
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);
```

```python
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);
ax.set_xlim([45,100]);
```

```python
## oo method from scratch

fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"])
## modify and style the graph
ax.set(title="Heart Disease and Cholestrol Levels", xlabel="Age", ylabel="Cholestrol")

## add a legend in the graph
ax.legend(*scatter.legend_elements(),title="Target")

## add a horizontal line in the graph

ax.axhline(over_50["chol"].mean(), linestyle="--");
```

```python
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_50["age"],
                   y=over_50["trestbps"],
                   c=over_50["target"])

```

```python
over_50.head()
```

```python
#subplots of chol, age and thalach

fig, (ax0,ax1) =plt.subplots(nrows=2, ncols=1, figsize=(10,10), sharex=True)

#adding data to our very first subplot
scatter1=ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);

# customizing our very first subplot
ax0.set(title="Heart Diease and Cholestrol Levels with Age",
        ylabel="Cholestrol");

#adding an index to the very first subplot
ax0.legend(*scatter.legend_elements(), title="Target");

#adding a mean line

ax0.axhline(over_50["chol"].mean(), linestyle="--");

## adding data to our second subplot
scatter2=ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"]);

## customizing our second subplot

ax1.set(title="Heart Disease and Max Heart Rate with Age",
        xlabel="Age",
        ylabel="Max Heart Rate");

#adding an index to the second plot
ax1.legend(*scatter.legend_elements(), title="Target");

#adding a mean line 
ax1.axhline(over_50["thalach"].mean(), linestyle="--");

## add a title to the figure

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
## Setting the style which we want to use

plt.style.use("seaborn-whitegrid")

## oo method from scratch

fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"],
                   cmap="plasma")
## modify and style the graph
ax.set(title="Heart Disease and Cholestrol Levels", xlabel="Age", ylabel="Cholestrol")

## add a legend in the graph
ax.legend(*scatter.legend_elements(),title="Target")

## add a horizontal line in the graph
ax.axhline(over_50["chol"].mean(),linestyle="--", color="black");
```

```python
#subplots of chol, age and thalach

fig, (ax0,ax1) =plt.subplots(nrows=2, ncols=1, figsize=(10,10), sharex=True)

#adding data to our very first subplot
scatter1=ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter");

# customizing our very first subplot
ax0.set(title="Heart Diease and Cholestrol Levels with Age",
        ylabel="Cholestrol");
#customizing the axes of ax0

ax0.set_xlim([50,80])

#adding an index to the very first subplot
ax0.legend(*scatter.legend_elements(), title="Target");

#adding a mean line

ax0.axhline(over_50["chol"].mean(), linestyle="--");

## adding data to our second subplot
scatter2=ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                     cmap="winter");

## customizing our second subplot

ax1.set(title="Heart Disease and Max Heart Rate with Age",
        xlabel="Age",
        ylabel="Max Heart Rate");

#changing the ax1 x limits
ax1.set_xlim([50,80])

#changing the limits of the ax1 y axis
ax1.set_ylim([60,200])

#adding an index to the second plot
ax1.legend(*scatter.legend_elements(), title="Target");

#adding a mean line 
ax1.axhline(over_50["thalach"].mean(), linestyle="--",color="black");

## add a title to the figure

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
